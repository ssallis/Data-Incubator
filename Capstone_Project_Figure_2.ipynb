```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from scipy import stats
pd.set_option('precision', 15)
```

```python
solitary = pd.read_csv('I:\\Data_Incubator\\icij-solitary-voices-final-dataset-for-publication.csv')
states = solitary['state'].unique()
average_hours_by_state = []
for state in states:
    average_hours_by_state.append([state,solitary['days_solitary'][solitary.state == state].sum()/solitary['days_solitary'][solitary.state == state].count()])
average_hours_by_state.sort(key=lambda x: x[0], reverse=False)
np.array(average_hours_by_state)[:,1]
```

```python
plt.figure(figsize=(22,18))
plt.bar(np.arange(len(np.array(average_hours_by_state))), np.array(np.array(average_hours_by_state)[:,1], dtype=float))
plt.xticks(np.arange(len(np.array(average_hours_by_state))), np.array(average_hours_by_state)[:,0], rotation='vertical')
plt.ylabel('Average Time Spent in Solitary')
plt.title('Time in Solitary by State')
```
